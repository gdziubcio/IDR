```python
import requests
import numpy as np
import matplotlib.pyplot as plt

def get_mobid_data(uniprot_id: str = 'P22980') -> dict:
    url = f"https://mobidb.org/api/download?acc={uniprot_id}&format=json"
    response = requests.get(url)
    protein_data: dict = response.json()
    return protein_data

def get_mobidb_fig(data) -> None:
    protein_length: int = data['length']
    protein_name: str = data['name']
    alphafold_scores: list[float] = data['prediction-disorder-alphafold']['scores']

    x = np.linspace(0, protein_length, len(alphafold_scores))
    y = np.array(alphafold_scores)

    fig, ax = plt.subplots()

    ax.plot(x, y, color='blue', label='AlphaFold Score')
    ax.set_xlim(0, protein_length)
    ax.set_xticks(np.arange(10, protein_length, 100))
    ax.set_ylim(0, 1)
    ax.set_yticks([0, 0.5, 1])

    # Hide the right and top spines
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)

    # Set only left and bottom spines to be visible over the data range
    ax.spines['left'].set_bounds(.5, 1)          # Y-axis bounds
    ax.spines['bottom'].set_bounds(0, protein_length)  # X-axis bounds

    # Set tick positions
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')

    plt.title(f"Protein: {protein_name} (Length: {protein_length})")
    plt.xlabel('Position')
    plt.ylabel('AlphaFold Scores')
    plt.legend()
    plt.show()

# Example usage
hlh_mobidb = get_mobid_data()
get_mobidb_fig(hlh_mobidb)
```
